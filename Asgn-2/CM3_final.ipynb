```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Data taken from CM1 and CM2

## Validation Accuracies of different algorithms on seeds dataset
acc_seed = [91.75, 91.75, 91.71]

## Validation Accuracies of different algorithms on covid dataset
acc_covid = [66.21, 66.88, 66.88]

## Test Accuracies of different algorithms on seeds dataset
acc_seed_test = [90, 90, 95]

## Test Accuracies of different algorithms on covid dataset
acc_covid_test = [66.72, 67.68, 66.99]

## Mean Fit time of different algorithms on seed dataset - as returned by GridSearchCV
time_seed = [9.8, 107.6, 176]

## Mean Fit time of different algorithms on covid dataset - as returned by GridSearchCV
time_covid = [93.8, 1177, 7212]

classifier = ['Decision Tree','Random Forest', 'Gradient Boosting']















```

### GRAPH OF VARIATION OF VALIDATION ACCURACY FOR DIFFERENT ALGORITHMS

```python
fig = plt.figure(figsize=(15,6))

plt.subplot(121)
plt.title("SEED DATASET VALIDATION ACCURACY")
sns.pointplot(x=classifier, y = acc_seed)
plt.xlabel("Classification Algorithm")
plt.ylabel("Accuracy %")

plt.subplot(122)
plt.title("COVID DATASET VALIDATION ACCURACY")
sns.pointplot(x=classifier, y = acc_covid)
plt.xlabel("Classification Algorithm")
plt.ylabel("Accuracy %")
```

### GRAPH OF VARIATION OF COMPUTATION TIME FOR DIFFERENT ALGORITHMS

```python
fig = plt.figure(figsize=(15,6))

plt.subplot(121)
plt.title("SEED DATASET COMPUTATION TIME")
sns.pointplot(x=classifier, y = time_seed)
plt.xlabel("Classification Algorithm")
plt.ylabel("Mean fit time--millisecond")

plt.subplot(122)
plt.title("COVID DATASET COMPUTATION TIME")
sns.pointplot(x=classifier, y = time_covid)
plt.xlabel("Classification Algorithm")
plt.ylabel("Mean fit time--millisecond")
```

### GRAPH OF VARIATION OF TEST ACCURACY FOR DIFFERENT ALGORITHMS

```python
fig = plt.figure(figsize=(15,6))

plt.subplot(121)
plt.title("SEED DATASET TEST ACCURACY")
sns.pointplot(x=classifier, y = acc_seed_test)
plt.xlabel("Classification Algorithm")
plt.ylabel("Accuracy %")

plt.subplot(122)
plt.title("COVID DATASET TEST ACCURACY")
sns.pointplot(x=classifier, y = acc_covid_test)
plt.xlabel("Classification Algorithm")
plt.ylabel("Accuracy %")
```

<p>The above results interprets:</p>
<ul>
            <li>The gradient Tree boosting outperformed decision tree classifier and random forest algorithm in case of the seed dataset. However, in case of covid dataset the random forest algorithm has outperformed decision tree algorithm and gradient tree boosting.</li>
            <li>Covid dataset is a complex dataset with mixture of nominal, ordinal and interval data. Moreover, the dataset was perpared using different encoding techniques and the final results shows that random forest classification algorithm has outperformed gradient boosting algorithm in terms of accuracy. The judgement of best performance cannot be made using the seed dataset as it has only numeric data and sometimes in case of a small dataset, gradient boosting may outperform random forest.</li>
    <li>
    Random Forests and Gradient Boosting result in a better accuracy over Decision Trees,as these algorithms help to reduce the bias and variance of final results. But this comes at a cost of additional computation time and complexity. Also these ensembles require more training data to perform better.
    </li>
            <li>The average computation time for Gradient tree boosting algorithm is very high compared to other two. On a large dataset it is almost exponential relation. It is one of the major draw back of this algorithm.</li>
            <li>None of the algorithms performed best on default parameters, which concludes that complex dataset requires parameter tuning.</li>
            <li>Despite the same validation accuracy, the test accuracy of random forest is higher than gradient boosting. This indicates overfitting of gradient tree boosting algorithm on the covid dataset.</li>
</ul>

<p>Is the performance difference what you expected given the algorithm and the datasets?</p>
<ul>
            <li>Before training the data, the expectation was that ensemble methods will outperform decision tree algorthim , which happened to be true. Since, the ensemble methods use bagging and boosting techiques to train the model in a better way improving after every split in case of boosting and for random forest averaging results obtained from random estimators.</li>
            <li>The expectation was that gradient boosting algorithm will out perform random forest, which was not the case in a relatively complex dataset (COVID).</li>
